### Data management

#### Finding null values

```python
sales.isnull().sum()
```

### Cleaning column names

```python
sales.columns = (
    sales.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False))
```

All the columns of the dataset were cleaned by removing whitespace, converting to lowercase, and replacing special characters for error-free and consistent manipulation.  

### Missing value treatment

```python
categorical_cols = ['gender', 'color', 'sleeve_type',
    'material', 'combo_item']
sales[categorical_cols] = sales[categorical_cols].fillna('Unknown')
```

There are several categorical columns containing missing values, and the missing values are replaced with “unknown” to avoid loss of informative data from the sales data. 

### Converting discount(%) to numeric

```python
sales['discount'] = (
    sales['discount%']
    .str.replace('%', '', regex=False)
    .astype(float))
sales.drop(columns=['discount%'], inplace=True)
```

```python
# Replacing missing discounts with 0
sales['discount'] = sales['discount'].fillna(0)
# dropping missing value product_name
sales = sales.dropna(subset=['product_name'])
```

In the case of the “discount%” column, the field was cleaned by removing the percentage symbol and converting the column to a numeric format. The null values of the column were replaced with 0, and the value indicates that no discount is applied to the product sales. 

### Checking duplicate values

```python
sales.duplicated().sum()
```

### Removing duplicate values

```python
sales = sales.drop_duplicates()
```

It has been seen that in the dataset, there are 1641 rows containing duplicate data based on the product sales. Duplicate records were dropped to prevent data bias issues in the collected sales data.

### Data wrangling

```python
# price bands
sales['price_band'] = pd.qcut(
    sales['price'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Premium'])
```

```python
# discount band
sales['discount_band'] = pd.qcut(
    sales['discount'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
# sales per review
sales['sales_per_review'] = (
    sales['no._of_sells'] / (sales['no._of_review'] + 1))
```

The data wrangling method is used to create categorical columns using quartile segmentation, and this section allows for making comparisons of product performance based on pricing levels. 

### Data analysis

```python
sales.groupby('gender')['no._of_sells'].mean().sort_values(ascending=False)
```

The data analysis highlights the mean product sales across different gender categories. The highest average sales were noticed in the female-targeted products, indicating that consumer demand isinfluenced by gender-specific marketing. Based on the dataset, relatively balanced market engagement was noticed across female and male segments. 

```python
sales.groupby('price_band')['no._of_sells'].mean()
```

It has been observed that high-priced products had thehighest average product sales, and the outcomes also suggest that medium and premium price category items have strong consumer demand. In this case, retailers can provide actionable insight based on pricing strategies, and value-driven pricing in premium and high segments can enhance the revenue of the business.

```python
sales.groupby('discount_band')['no._of_sells'].mean()
```

The outcomes demonstrate that a high discount band helped in producing the highest average sales, as well as low and medium discounts on the products decrease the product sales. Thus, it can be said that large discounts on products enhance product demand and influence consumer purchasing behaviour. Retailers can use promotional strategies to improve market competitiveness and enhance the sales volume of their products. 

```python
sales.groupby('location')['no._of_sells'].mean()
```

Rangpur has the highest average sales among the other locations of Bangladesh.        

### Data visualisation

```python
# Price vs number of Sales
plt.figure(figsize=(8,5))
sns.scatterplot(x=sales['price'], y=sales['no._of_sells'], alpha=0.5)
plt.title('Price vs Number of Sales')
plt.xlabel('Price')
plt.ylabel('Number of Sales')
plt.show()
```

The scatter plot demonstrates the way product prices relate to the number of product sales. It has been observed that most products are concentrated in the lower price range. An inverse relationship was noticed here between price and the number of product sales. 

```python
# Finding top 10 products
top10_price = (
    sales[['product_name', 'price', 'no._of_sells', 'discount']]
    .sort_values(by='price', ascending=False)
    .head(10))
```

```python
# top 10 most expensive products 
plt.figure(figsize=(10,6))
sns.barplot(data=top10_price, y='product_name', x='price')
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Price')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

The above bar chart is developed here to identify the most expensive products on an e-commerce platform. The chart shows that branded and premium T-shirts are recorded in the upper price range, and the high-priced products of the online platform reflect good product value and quality. The data visualisation offers actionable insights into addressing targeted marketing and luxury product categories to enhance the profitability of the retailers. 

```python
#distribution of product price
plt.figure(figsize=(8,5))
plt.hist(sales['price'], bins=30)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

The above histogram plot explores the product price distribution based on the sales dataset of Daraz. It has been noticed that most of the products on the online platform fall in the lower price range, indicating that affordable items aredominated in the e-commerce market. Lower-cost products help in enhancing the sales volume of the e-commerce platform. 

```python
#customer review vs sales
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=sales['no._of_review'],
    y=sales['no._of_sells'],
    color='green',
    alpha=0.6)
plt.title('Number of Reviews vs Number of Sales')
plt.xlabel('Number of Reviews')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.show()
```

The scatter plot highlights a strong relationship between the number of customer reviews and the number of product sales. Products with more customer reviews help in enhancing the sales of the products, and in this case, customer engagement plays a vital role in influencing the purchasing decisions of consumers. Customer reviews on products help in increasing sales growth, improving product performance, and building customer trust in the e-commerce market.

### Data visualisation

```python
#import and export trend over year
plt.figure(figsize=(8,5))
plt.plot(
    bangladesh_lpi['Year'],
    bangladesh_lpi['Import'],
    marker='o',
    label='Import')
plt.plot(
    bangladesh_lpi['Year'],
    bangladesh_lpi['Export'],
    marker='o',
    label='Export')
plt.title('Bangladesh Import and Export Trends')
plt.xlabel('Year')
plt.ylabel('Trade Value')
plt.legend()
plt.tight_layout()
plt.show()
```

The above line chart demonstrates the trend of import and export value of Bangladesh over time. The import and export trade volume shows upward growth over time. In such circumstances, enhancing economic activity increases the marketing demand for the products. Rising import and export business performance influences product availability and sales performance in the country. 

```python
#trade balance over time
plt.figure(figsize=(8,5))
sns.lineplot(
    x=bangladesh_lpi['Year'],
    y=bangladesh_lpi['Trade Balance'],
    marker='o',
    color='blue')
plt.title('Trade Balance Trend in Bangladesh')
plt.xlabel('Year')
plt.ylabel('Trade Balance')
plt.tight_layout()
plt.show()
```

The line chart shows the trade balance of Bangladesh over the years. A slight improvement was noticed between 2014 and 2016, and the trend highlights the enhancement of reliance on imported products on the online platform. Imbalanced trade affects the purchasing patterns of customers and economic stability. 

```python
#Gdp per capita analysis
plt.figure(figsize=(8,5))
sns.lineplot(
    x=bangladesh_lpi['Year'],
    y=bangladesh_lpi['gdp_per_capita'],
    marker='o',
    color='purple')
plt.title('GDP per Capita Trend in Bangladesh')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.tight_layout()
plt.show()
```

The line chart shows a steady rise in GDP per capita of Bangladesh, and the data visualisation shows higher average income levels and economic improvement. An upward trend was noticed in analysing the economic condition of Bangladesh, and the growth indicates strong market demand. Based on the chart, it has been noticed that the income levels of consumers increases and this factor positively influences digital market expansion and e-commerce sales of Bangladesh.
